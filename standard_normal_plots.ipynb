```python
class Vector(object):
  def __init__(self,x=0, y= 0, z = 0):
    self.x= x
    self.y = y
    self.z = z

v = Vector(1,2,3)
```

```python
class Vector(Vector):
  def __repr__(self):
    return "Vector(%r,%r,%r)" % (self.x,self.y,self.z)

v =Vector(1,2,3)
v
```

```python
class Vector(Vector):
  def __abs__(self):
    return (self.x**2 + self.y**2 + self.z**2)**0.5

  def __bool__(self):
    return bool(abs(self))

v = Vector(1,2,-1)
abs(v)

bool(v)
v = Vector()
v
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style = "darkgrid")
mpl.rcParams["font.family"] = "serif"
%matplotlib inline
```

```python
np.random.seed(1000)
y = np.random.standard_normal(20)
x = np.arange(len(y))

plt.figure(figsize = (10,6))
plt.plot(y.cumsum(), "m",lw =1.5)
plt.plot(y.cumsum(),"ro")
plt.xlabel('index')
plt.ylabel("value")
plt.title("A Simple Plot");
```

```python
y = np.random.standard_normal((20,2)).cumsum(axis=0)
plt.figure(figsize =(10,6))
plt.subplot(121)
plt.plot(y[:,0],"r", lw =1.5, label ="1st")
plt.plot(y[:,0],"b")
plt.legend(loc = "best")
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title("A Simple Plot")
plt.subplot(122)
plt.bar(np.arange(len(y)), y[:,1], width = 0.5, color = "g", label = "2nd")
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value');
```

```python
def func(x):
  return 0.5 *np.exp(x) +1
a ,b =0.5,1.5
x = np.linspace(0,2)
y = func(x)
Ix = np.linspace(a,b)
Iy = func(Ix)
verts = [(a,0)] + list(zip(Ix,Iy))+[(b,0)]
```

```python
from matplotlib.patches import Polygon
fig, ax = plt.subplots(figsize =(10,6))
plt.plot(x,y,"b", linewidth =2)
plt.ylim(bottom = 0)
poly = Polygon(verts, facecolor = "0.7",edgecolor ="0.5")
ax.add_patch(poly)
plt.text(0.5 *(a+b), 1, r'$\int_a^b f(x)\mathrm{d}x$', horizontalalignment = "center", fontsize =20)
plt.figtext(0.9,.075, "$b$")
plt.figtext(0.075,.9, "$f(x)$")
ax.set_xticks((a,b))
ax.set_xticklabels(("$a$","$b$"))
ax.set_yticks([func(a),func(b)])
ax.set_yticklabels(("$f(a)$","$f(b)$"))

```

```python
strike = np.linspace(50,150,24)
ttm = np.linspace(.5,2.5,24)
strike, ttm = np.meshgrid(strike,ttm)
strike[:2].round(1)
iv = (strike -100)**2 / (100 * strike) / ttm
iv[:5,:3]
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,6))
ax =fig.add_subplot(111,projection = "3d")
surf = ax.plot_surface(strike,ttm,iv,rstride =2, cstride =2, cmap= plt.cm.coolwarm,
                       linewidth =0.5, antialiased = True)

ax.set_xlabel("strike")
ax.set_ylabel("maturity")
ax.set_zlabel("implied volatility")
fig.colorbar(surf,shrink =0.5, aspect = 0.5)
```

```python
import random
def average_py(n):
  s=0
  for i in range(n):
    s += random.random()
  return s/n

n = 10000000
%time average_py(n)
%timeit average_py(n)
```

```python
import numba
average_nb = numba.jit(average_py)
%timeit average_nb(n)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


S0 =100
r= .05
sigma =0.25
T = 30 / 365.
I =10000
rng = np.random.default_rng()
ST = S0 * np.exp((r-0.5 *sigma **2) *T + sigma *np.sqrt(T) * rng.standard_normal(I))

R_gbm = np.sort(ST -S0)

plt.figure(figsize=(10,6))
plt.hist(R_gbm, bins =50)
plt.xlabel('abs return')
plt.ylabel('freq')
```
